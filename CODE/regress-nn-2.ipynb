```python
from sklearn.feature_selection import SelectKBest, r_regression,f_regression,mutual_info_regression
skb = SelectKBest(score_func=r_regression, k=40)
skb.fit(X,y)
X_selected=skb.transform(X)
build_test_selected=skb.transform(build_test)
n_features=X_selected.shape[1]
print(n_features)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.2, random_state=0)
```

```python
# pip install torch
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.regularizers import L1,L1L2,L2
from keras.callbacks import EarlyStopping
from keras.optimizers import AdamW, SGD, RMSprop, Adadelta, Adagrad, Adamax,Adafactor, Nadam, Ftrl
from keras.metrics import RootMeanSquaredError
import torch
os.environ["KERAS_BACKEND"] = "torch"
from keras.optimizers.schedules import InverseTimeDecay
```

```python

# from keras.wrappers.scikit_learn import KerasRegressor


model=Sequential()
# model.add(Dense(100, input_dim=15,activation='relu'))
# model.add(Dense(50, activation='relu'))
# model.add(Dense(25, activation='relu'))
# model.add(Dense(10, activation='relu'))

model.add(Dense(256, input_dim=n_features,activation='relu',kernel_regularizer=L1()))
# model.add(Dropout(0.3))
model.add(Dense(128, activation='sigmoid', kernel_regularizer=L1()))
# model.add(Dropout(0.25))
model.add(Dense(64, activation='relu', kernel_regularizer=L1()))
# model.add(Dropout(0.2))
model.add(Dense(32, activation='sigmoid', kernel_regularizer=L1()))
model.add(Dense(16, activation='relu', kernel_regularizer=L1()))
model.add(Dense(8, activation='relu', kernel_regularizer=L1()))
model.add(Dense(4, activation='relu', kernel_regularizer=L1()))
# model.add(Dense(2, activation='relu'))
# Output layer
model.add(Dense(1, activation='linear',kernel_regularizer=L1()))

# Compile the model
learning_rate_fn=InverseTimeDecay(
  0.001,
  decay_steps=284*1000,
  decay_rate=1,
  staircase=False)
opt = Adamax(learning_rate=learning_rate_fn)
model.compile(optimizer=opt, loss='mean_squared_error', metrics=[RootMeanSquaredError()])




model.summary()
model.fit(X_train, y_train, epochs=100, batch_size=512, callbacks=EarlyStopping(monitor='loss', patience=10, min_delta=100000000))
model.evaluate(X_test, y_test, batch_size=4096)
```
