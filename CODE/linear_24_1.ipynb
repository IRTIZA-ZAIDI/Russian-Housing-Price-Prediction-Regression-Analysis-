```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
import csv

# Load the training data
df = pd.read_csv('train.csv')

# Preprocess the data (same as before)
category = df['sub_area'].value_counts()
others_col = category[category < 600].index.tolist()
df['sub_area'] = df['sub_area'].replace(others_col, 'others')
df = pd.get_dummies(df, drop_first=True)

# Split into features (X) and target variable (y)
X = df.loc[:, df.columns != 'price_doc']
y = df[['price_doc']]

# Feature selection
selector = SelectKBest(f_regression, k=30)
X_selected = selector.fit_transform(X, y)

# Load the test data
df_test = pd.read_csv('test.csv')
df_test = df_test.drop(['row ID'], axis=1)
df_test['sub_area'] = df_test['sub_area'].replace(others_col, 'others')
df_test = pd.get_dummies(df_test, drop_first=True)

# Apply the same feature selection to the test data
df_test_selected = selector.transform(df_test)

# Train the linear regression model on the selected features
reg = LinearRegression().fit(X_selected, y)

# Make predictions on the test data
y_pred_test = reg.predict(df_test_selected)

# Save predictions to a CSV file
filepath = 'prediction_linear_24nov_selected_features.csv'
with open(filepath, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['row ID', 'price_doc'])  # Add column headers
    for c, i in enumerate(y_pred_test, start=1):
        writer.writerow([c, i[0]])  # Remove square brackets around the predicted value.
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import csv

# Load the training data
df = pd.read_csv('train.csv')

# Preprocess the data (same as before)
category = df['sub_area'].value_counts()
others_col = category[category < 600].index.tolist()
df['sub_area'] = df['sub_area'].replace(others_col, 'others')
df = pd.get_dummies(df, drop_first=True)

# Split into features (X) and target variable (y)
X = df.loc[:, df.columns != 'price_doc']
y = df[['price_doc']]

# Apply PCA for 50 components
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X)

# Load the test data
df_test = pd.read_csv('test.csv')
df_test = df_test.drop(['row ID'], axis=1)
df_test['sub_area'] = df_test['sub_area'].replace(others_col, 'others')
df_test = pd.get_dummies(df_test, drop_first=True)

# Apply the same PCA to the test data
df_test_pca = pca.transform(df_test)

# Train the linear regression model on the PCA-transformed features
reg = LinearRegression().fit(X_pca, y)

# Make predictions on the test data
y_pred_test = reg.predict(df_test_pca)

# Save predictions to a CSV file
filepath = 'prediction_linear_23nov_pca.csv'
with open(filepath, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['row ID', 'price_doc'])  # Add column headers
    for c, i in enumerate(y_pred_test, start=1):
        writer.writerow([c, i[0]])  # Remove square brackets around the predicted value.
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler
import csv

# Load the training data
df = pd.read_csv('train.csv')

# Preprocess the data (same as before)
category = df['sub_area'].value_counts()
others_col = category[category < 600].index.tolist()
df['sub_area'] = df['sub_area'].replace(others_col, 'others')
df = pd.get_dummies(df, drop_first=True)

# Split into features (X) and target variable (y)
X = df.loc[:, df.columns != 'price_doc']
y = df[['price_doc']]

# Apply VarianceThreshold to remove low-variance features
variance_threshold = VarianceThreshold(threshold=0.05)
X_filtered = variance_threshold.fit_transform(X)

# Apply SelectKBest with f_regression to select top 20 features
k_best = SelectKBest(f_regression, k=15)
X_selected = k_best.fit_transform(X_filtered, y)

# Train the linear regression model on the selected features
reg = LinearRegression().fit(X_selected, y)

# Load the test data
df_test = pd.read_csv('test.csv')
df_test = df_test.drop(['row ID'], axis=1)
df_test['sub_area'] = df_test['sub_area'].replace(others_col, 'others')
df_test = pd.get_dummies(df_test, drop_first=True)

# Apply the same feature selection to the test data
df_test_filtered = variance_threshold.transform(df_test)
df_test_selected = k_best.transform(df_test_filtered)

# Make predictions on the test data
y_pred_test = reg.predict(df_test_selected)

# Save predictions to a CSV file
filepath = 'prediction_linear_23nov_filtered_all_test.csv'
with open(filepath, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['row ID', 'price_doc'])  # Add column headers
    for c, i in enumerate(y_pred_test, start=1):
        writer.writerow([c, i[0]])  # Remove square brackets around the predicted value.
```
